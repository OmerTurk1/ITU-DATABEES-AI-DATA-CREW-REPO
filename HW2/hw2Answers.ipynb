## Section 2: Data Filtering and Grouping

### Task 1: Assume that students who studies over 25 hours are "hardworking", while those who studies less than 15 hours are "lazy". What is the averages of previous exam scores for these groups?

```python
hardworking = df[df['Hours_Studied'] > 25]
lazy = df[df['Hours_Studied'] < 15]
hw_avg = hardworking['Previous_Scores'].mean()
lz_avg = lazy['Previous_Scores'].mean()
print("hardworking average:",hw_avg)
print("lazy average:",lz_avg)
```
